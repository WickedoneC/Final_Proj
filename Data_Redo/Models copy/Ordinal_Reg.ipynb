```python
import mord
from sklearn import linear_model, metrics, preprocessing
```

```python
clf2 = mord.LogisticAT(alpha=1)
clf2.fit(X, newy)
print('Mean Absolute Error of LogisticAT %s' %
      metrics.mean_absolute_error(clf2.predict(X), newy))
```

```python
clf3 = mord.LogisticIT(alpha=6, verbose=100, max_iter=1000000)
clf3.fit(X, newy)
print('Mean Absolute Error of LogisticIT %s' %
      metrics.mean_absolute_error(clf3.predict(X), newy))
```

```python
# Actual class predictions
rf_predictions = clf3.predict(X)
# Probabilities for each class
rf_probs = clf3.predict_proba(X)[:, 1]
```

```python
clf4 = mord.LogisticSE(alpha=10.)
clf4.fit(X, newy)
print('Mean Absolute Error of LogisticSE %s' %
      metrics.mean_absolute_error(clf4.predict(X), newy))
```

```python
clf1 = linear_model.LogisticRegression(
    solver='lbfgs',
    multi_class='multinomial')
clf1.fit(X, newy)

print('Mean Absolute Error of LogisticRegression: %s' %
      metrics.mean_absolute_error(clf1.predict(X), newy))
```
