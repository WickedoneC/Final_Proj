```python
k_means = KMeans(n_clusters = 15)
k_means.fit(Z)
z_hat = k_means.predict(Z)
```

```python
plt.scatter(Z[:, 0], Z[:, 1], c = z_hat, s = 25)
cl_centers = k_means.cluster_centers_
plt.scatter(cl_centers[:, 0], cl_centers[:, 1], c='black', s=100);
```

```python
from sklearn import metrics
labels = k_means.labels_

metrics.silhouette_score(Z, labels, metric='euclidean')
```

```python
k_means_6 = KMeans(n_clusters = 16)
k_means_6.fit(Z)
zhat_6 = k_means_6.predict(Z)
plt.scatter(Z[:, 0], Z[:, 1], c = zhat_6, s = 25)
cl_centers_6 = k_means_6.cluster_centers_
plt.scatter(cl_centers_6[:, 0], cl_centers_6[:, 1], c='black', s=100);
```
